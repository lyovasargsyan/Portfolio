```python
data_info["Pass_id"] = "_"
x_train["PassengerId1"] = x_train["PassengerId"].str.split(data_info["Pass_id"]).str[0]
x_train["PassengerId2"] = x_train["PassengerId"].str.split(data_info["Pass_id"]).str[1]
x_train.drop(columns="PassengerId", inplace = True)
```

```python
data_info["HomePlanet_counts"] = x_train.HomePlanet.value_counts() / x_train.HomePlanet.value_counts().sum()
x_train["HomePlanet"] = x_train["HomePlanet"].apply(
    lambda x: x if x in data_info["HomePlanet_counts"].keys() \
              else np.random.choice(data_info["HomePlanet_counts"].keys(), p=data_info["HomePlanet_counts"].values))
for i in data_info["HomePlanet_counts"].keys():
    x_train[f"HomePlanet_{i}"] = (x_train["HomePlanet"] == i).astype("int8")
x_train.drop("HomePlanet", axis=1, inplace=True)
```

```python
data_info['CryoSleep_mode'] = False
x_train["CryoSleep"] = (x_train["CryoSleep"].fillna(data_info['CryoSleep_mode']))
x_train["CryoSleep"] = x_train["CryoSleep"].apply(lambda x: 1 if x==True else 0)
```

```python
data_info["Age_counts"] = x_train.Age.value_counts() / x_train.Age.value_counts().sum()
x_train["Age"] = x_train["Age"].apply(
    lambda x: x if x in data_info["Age_counts"].keys() \
              else np.random.choice(data_info["Age_counts"].keys(), p=data_info["Age_counts"].values))
```

```python
data_info["Cabin_filna"] = "0/0/0"
x_train.Cabin = x_train.Cabin.fillna(data_info["Cabin_filna"])

data_info["Cabin_split"] = "/"

x_train["Cabin1"] = x_train["Cabin"].str.split(data_info["Cabin_split"]).str[0]
x_train["Cabin1"] = x_train["Cabin1"].apply(lambda x: np.nan if x == "0" else x)
data_info["Cabin1_counts"] = x_train["Cabin1"].value_counts()[:7] / x_train["Cabin1"].value_counts()[:7].sum()
x_train["Cabin1"] = x_train["Cabin1"].apply(
    lambda x: x if x in data_info["Cabin1_counts"].keys() \
              else np.random.choice(data_info["Cabin1_counts"].keys(), p=data_info["Cabin1_counts"].values))
for i in data_info["Cabin1_counts"].keys():
    x_train[f"Cabin1_{i}"] = (x_train["Cabin1"] == i).astype("int8")
x_train.drop("Cabin1", axis=1, inplace=True)

x_train["Cabin2"] = x_train["Cabin"].str.split(data_info["Cabin_split"]).str[1]
x_train["Cabin2"] = x_train["Cabin2"].apply(lambda x: np.nan if x == "0" else x)
data_info["Cabin2_counts"] = x_train["Cabin2"].value_counts() / x_train["Cabin2"].value_counts().sum()
x_train["Cabin2"] = x_train["Cabin2"].apply(
    lambda x: x if x in data_info["Cabin2_counts"].keys() \
              else np.random.choice(data_info["Cabin2_counts"].keys(), p=data_info["Cabin2_counts"].values))

x_train["Cabin3"] = x_train["Cabin"].str.split(data_info["Cabin_split"]).str[2]
x_train["Cabin3"] = x_train["Cabin3"].apply(lambda x: np.nan if x == "0" else x)
data_info["Cabin3_counts"] = x_train["Cabin3"].value_counts() / x_train["Cabin3"].value_counts().sum()
x_train["Cabin3"] = x_train["Cabin3"].apply(
    lambda x: x if x in data_info["Cabin3_counts"].keys() \
              else np.random.choice(data_info["Cabin3_counts"].keys(), p=data_info["Cabin3_counts"].values))
x_train["Cabin3"] = x_train["Cabin3"].apply(lambda x: 1 if x=="S" else 0)

x_train.drop(columns="Cabin", inplace = True)
```

```python
data_info["Destination_counts"] = x_train.Destination.value_counts() / x_train.Destination.value_counts().sum()
x_train["Destination"] = x_train["Destination"].apply(
    lambda x: x if x in data_info["Destination_counts"].keys() \
              else np.random.choice(data_info["Destination_counts"].keys(), p=data_info["Destination_counts"].values))
for i in data_info["Destination_counts"].keys():
    x_train[f"Destination_{i}"] = (x_train["Destination"] == i).astype("int8")
x_train.drop("Destination", axis=1, inplace=True)
```

```python
data_info['VIP_mode'] = False
x_train["VIP"] = (x_train["VIP"].fillna(data_info['VIP_mode']))
x_train["VIP"] = x_train["VIP"].apply(lambda x: 1 if x==True else 0)
```

```python
data_info['RoomService_mode'] = 0
x_train["RoomService"] = (x_train["RoomService"].fillna(data_info['RoomService_mode']))
```

```python
data_info['FoodCourt_mode'] = 0
x_train["FoodCourt"] = (x_train["FoodCourt"].fillna(data_info['FoodCourt_mode']))
```

```python
data_info['ShoppingMall_mode'] = 0
x_train["ShoppingMall"] = (x_train["ShoppingMall"].fillna(data_info['ShoppingMall_mode']))
```

```python
data_info['Spa_mode'] = 0
x_train["Spa"] = (x_train["Spa"].fillna(data_info['Spa_mode']))
```

```python
data_info['VRDeck_mode'] = 0
x_train["VRDeck"] = (x_train["VRDeck"].fillna(data_info['VRDeck_mode']))
```

```python
x_train.drop(columns="Name", inplace=True)
```

```python
def data_analysis(df, data_info):
    df["PassengerId1"] = df["PassengerId"].str.split(data_info["Pass_id"]).str[0]
    df["PassengerId2"] = df["PassengerId"].str.split(data_info["Pass_id"]).str[1]
    df.drop(columns="PassengerId", inplace = True)


    
    df["HomePlanet"] = df["HomePlanet"].apply(
        lambda x: x if x in data_info["HomePlanet_counts"].keys() \
                  else np.random.choice(data_info["HomePlanet_counts"].keys(), p=data_info["HomePlanet_counts"].values))
    for i in data_info["HomePlanet_counts"].keys():
        df[f"HomePlanet_{i}"] = (df["HomePlanet"] == i).astype("int8")
    df.drop("HomePlanet", axis=1, inplace=True)


    
    df["CryoSleep"] = (df["CryoSleep"].fillna(data_info['CryoSleep_mode']))
    df["CryoSleep"] = df["CryoSleep"].apply(lambda x: 1 if x==True else 0)
    
    df["Age"] = df["Age"].apply(
    lambda x: x if x in data_info["Age_counts"].keys() \
              else np.random.choice(data_info["Age_counts"].keys(), p=data_info["Age_counts"].values))
    
    df.Cabin = df.Cabin.fillna(data_info["Cabin_filna"])
    
    df["Cabin1"] = df["Cabin"].str.split(data_info["Cabin_split"]).str[0]
    df["Cabin1"] = df["Cabin1"].apply(lambda x: np.nan if x == "0" else x)
    df["Cabin1"] = df["Cabin1"].apply(
        lambda x: x if x in data_info["Cabin1_counts"].keys() \
                  else np.random.choice(data_info["Cabin1_counts"].keys(), p=data_info["Cabin1_counts"].values))
    for i in data_info["Cabin1_counts"].keys():
        df[f"Cabin1_{i}"] = (df["Cabin1"] == i).astype("int8")
    df.drop("Cabin1", axis=1, inplace=True)
    
    df["Cabin2"] = df["Cabin"].str.split(data_info["Cabin_split"]).str[1]
    df["Cabin2"] = df["Cabin2"].apply(lambda x: np.nan if x == "0" else x)
    df["Cabin2"] = df["Cabin2"].apply(
        lambda x: x if x in data_info["Cabin2_counts"].keys() \
                  else np.random.choice(data_info["Cabin2_counts"].keys(), p=data_info["Cabin2_counts"].values))
    
    df["Cabin3"] = df["Cabin"].str.split(data_info["Cabin_split"]).str[2]
    df["Cabin3"] = df["Cabin3"].apply(lambda x: np.nan if x == "0" else x)
    df["Cabin3"] = df["Cabin3"].apply(
        lambda x: x if x in data_info["Cabin3_counts"].keys() \
                  else np.random.choice(data_info["Cabin3_counts"].keys(), p=data_info["Cabin3_counts"].values))
    df["Cabin3"] = df["Cabin3"].apply(lambda x: 1 if x=="S" else 0)
    
    df.drop(columns="Cabin", inplace = True)


    
    df["Destination"] = df["Destination"].apply(
        lambda x: x if x in data_info["Destination_counts"].keys() \
                  else np.random.choice(data_info["Destination_counts"].keys(), p=data_info["Destination_counts"].values))
    for i in data_info["Destination_counts"].keys():
        df[f"Destination_{i}"] = (df["Destination"] == i).astype("int8")
    df.drop("Destination", axis=1, inplace=True)
    
    
    
    df["VIP"] = (df["VIP"].fillna(data_info['VIP_mode']))
    df["VIP"] = df["VIP"].apply(lambda x: 1 if x==True else 0)
    
    
    df["RoomService"] = (df["RoomService"].fillna(data_info['RoomService_mode']))
    
    
    df["FoodCourt"] = (df["FoodCourt"].fillna(data_info['FoodCourt_mode']))
    
    
    df["ShoppingMall"] = (df["ShoppingMall"].fillna(data_info['ShoppingMall_mode']))
    
    
    df["Spa"] = (df["Spa"].fillna(data_info['Spa_mode']))
    
    
    df["VRDeck"] = (df["VRDeck"].fillna(data_info['VRDeck_mode']))

    
    df.drop(columns="Name", inplace=True)

    return df
```
