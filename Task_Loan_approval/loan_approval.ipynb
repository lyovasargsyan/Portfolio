```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from collections import Counter
```

```python
data = pd.read_csv("loan_approval.csv")
```

```python
data
```

```python
data.info()
```

```python
data = data.drop(columns = ["name", "city"])
```

```python
y = data.loan_approved.values
x = data.drop(columns = "loan_approved")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, shuffle=True, test_size=0.2, random_state=42)
```

```python
clf = DecisionTreeClassifier(max_depth=5, min_samples_leaf=3, min_samples_split=2, criterion='gini', random_state=42)
clf.fit(x_train, y_train)
print(f'Test accuracy: {accuracy_score(y_test, clf.predict(x_test))}')
print(f'Train accuracy: {accuracy_score(y_train, clf.predict(x_train))}')
```

```python
fig = plt.figure(figsize=(5, 5))
tree.plot_tree(clf, feature_names=data.columns, filled=True)
```
