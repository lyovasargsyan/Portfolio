```python
import numpy as np
import pandas as pd
import matplotlib as plt 
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('taxi_trip_pricing.csv')
```

```python
data
```

```python
data["Per_Km_Rate"] = data["Per_Km_Rate"].fillna(data["Per_Km_Rate"].mean())
```

```python
data.Day_of_Week.fillna("Weekday")
data.Day_of_Week = (data["Day_of_Week"] == "Weekday").astype("int8")
```

```python
data["Time_of_Day"] = data["Time_of_Day"].fillna(data["Time_of_Day"].mode()[0])
data = pd.get_dummies(data, columns=["Time_of_Day"], dtype = "int8")
```

```python
data.Trip_Distance_km = data.Trip_Distance_km.fillna(data.Trip_Distance_km.mean())
```

```python
data["Traffic_Conditions"] = data["Traffic_Conditions"].fillna(data["Traffic_Conditions"].mode()[0])
data = pd.get_dummies(data, columns=["Traffic_Conditions"], dtype = "int8")
```

```python
data["Weather"] = data["Weather"].fillna(data["Weather"].mode()[0])
data = pd.get_dummies(data, columns=["Weather"], dtype = "int8")
```

```python
data["Base_Fare"] = data["Base_Fare"].fillna(data["Base_Fare"].mode()[0])
```

```python
data["Per_Minute_Rate"] = data["Per_Minute_Rate"].fillna(data["Per_Minute_Rate"].mean())
```

```python
data["Trip_Duration_Minutes"] = data["Trip_Duration_Minutes"].fillna(data["Trip_Duration_Minutes"].mean())
```

```python
data["Trip_Price"] = data["Trip_Price"].fillna(data["Trip_Price"].mean())
```

```python
data["Passenger_Count"] = data["Passenger_Count"].fillna(data["Passenger_Count"].mean().round()).astype('int8')
```

```python
data.info()
```

```python
y = data.Trip_Price.values
x = data.drop("Trip_Price", axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(x_train)
X_test_poly = poly.fit_transform(x_test)
```

```python
model = LinearRegression()
model.fit(X_poly, y_train)
```

```python
y_pred = model.predict(X_test_poly)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mse, r2
```
