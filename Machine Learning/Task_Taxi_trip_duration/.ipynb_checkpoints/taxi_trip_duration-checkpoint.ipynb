```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import accuracy_score

from xgboost import XGBRegressor
```

```python
train = pd.read_csv("train.csv")
```

```python
test = pd.read_csv("test.csv")
```

```python
res = pd.read_csv("sample_submission.csv")
```

```python
train.drop("id", axis=1, inplace=True)
```

```python
train.store_and_fwd_flag = (train.store_and_fwd_flag == "N").astype("int8")
```

```python
train = pd.get_dummies(train, columns=["vendor_id"], dtype="int8")
```

```python
train["pickup_datetime"] = pd.to_datetime(train["pickup_datetime"])
train["dropoff_datetime"] = pd.to_datetime(train["dropoff_datetime"])
```

```python
train["duration"] = (train["dropoff_datetime"] - train["pickup_datetime"]).dt.seconds
```

```python
train.drop(index=train[(train["trip_duration"] != train["duration"])].index, inplace=True)
```

```python
train.drop(columns=["dropoff_datetime", "duration"], inplace=True)
```

```python
train["year"] = train["pickup_datetime"].dt.year
train["month"] = train["pickup_datetime"].dt.month
train["day"] = train["pickup_datetime"].dt.day
train["hour"] = train["pickup_datetime"].dt.hour
train["weekday"] = train["pickup_datetime"].dt.weekday
```

```python
train = train.drop("pickup_datetime", axis=1)
```

```python
train["distance"] = np.sqrt(
    (train["dropoff_longitude"] - train["pickup_longitude"])**2 +
    (train["dropoff_latitude"] - train["pickup_latitude"])**2
)
```

```python
train.hist(bins = 20)
```

```python
def rmsle(y_true, y_pred):
    y_pred = tf.maximum(y_pred, 0.0) 

    log_true = tf.math.log1p(y_true)
    log_pred = tf.math.log1p(y_pred)

    return tf.sqrt(tf.reduce_mean(tf.square(log_true - log_pred)))
```

```python
x = train.drop(columns="trip_duration")
```

```python
y = train["trip_duration"].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = tf.keras.Sequential([
    Input(x_train.shape[1:], 32),
    Dense(1000, activation='sigmoid'),
    Dense(1000, activation='relu'),
    Dense(1)
])
model.summary()
```

```python
model.compile(optimizer='adam',
    loss='mse',
    metrics=[rmsle])
```

```python
model.fit(x_train, y_train, 32, 10, validation_data=(x_test, y_test))
```
